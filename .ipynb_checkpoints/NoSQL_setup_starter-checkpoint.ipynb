1. An exciting new halal restaurant just opened in Greenwich, but hasn't been rated yet. The magazine has asked you to include it in your analysis. Add the following restaurant "Penang Flavours" to the database.

```python
# Create a dictionary for the new restaurant data
# Create a dictionary for the new restaurant data
new_restaurant = {
    'BusinessName': 'Penang Flavours',
    'BusinessType': 'Restaurant',
    'BusinessTypeID': 1234,
    'AddressLine1': '123 Main Street',
    'AddressLine2': 'City',
    'AddressLine3': 'State',
    'AddressLine4': '',
    'PostCode': '12345',
    'Phone': '',
    'RatingValue': '',
    'RatingKey': '',
    'RatingDate': '',
    'LocalAuthorityCode': '456',
    'LocalAuthorityName': 'Local Authority',
    'LocalAuthorityWebSite': 'https://example.com',
    'LocalAuthorityEmailAddress': 'info@example.com',
    'scores': {
        'Hygiene': 0,
        'Structural': 0,
        'ConfidenceInManagement': 0
    },
    'SchemeType': 'FHRS',
    'geocode': {
        'longitude': '0.0',
        'latitude': '0.0'
    },
    'RightToReply': '',
    'Distance': 0.0,
    'NewRatingPending': False
}
```

```python
# Insert the new restaurant into the collection
result = establishments.insert_one(new_restaurant)
```

```python
# Check that the new restaurant was inserted
# Find the newly inserted restaurant by name
inserted_restaurant = establishments.find_one({ 'BusinessName': 'Penang Flavours' })

# Print the inserted restaurant document
print(inserted_restaurant)
```

2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
# Define the query
query = { 'BusinessType': 'Restaurant/Cafe/Canteen' }

# Define the projection to include only the BusinessTypeID and BusinessType fields
projection = { '_id': 0, 'BusinessTypeID': 1, 'BusinessType': 1 }

# Execute the query
restaurants = establishments.find(query, projection)

# Print the results
for restaurant in restaurants:
    print(restaurant)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID
filter = { 'BusinessName': 'Penang Flavours' }

# Define the update to set the BusinessTypeID to 1
update = { '$set': { 'BusinessTypeID': 1 } }

# Update the restaurant document
result = establishments.update_one(filter, update)
```

```python
# Confirm that the new restaurant was updated
# Find the updated restaurant by name
updated_restaurant = establishments.find_one({ 'BusinessName': 'Penang Flavours' })

# Print the updated restaurant document
print(updated_restaurant)
```

4. The magazine is not interested in any establishments in Dover, so check how many documents contain the Dover Local Authority. Then, remove any establishments within the Dover Local Authority from the database, and check the number of documents to ensure they were deleted.

```python
# Find how many documents have LocalAuthorityName as "Dover"

query = { 'LocalAuthorityName': 'Dover' }


count = establishments.count_documents(query)


print(f"Number of documents with LocalAuthorityName 'Dover': {count}")
```

```python
# Delete all documents where LocalAuthorityName is "Dover"
query = { 'LocalAuthorityName': 'Dover' }


result = establishments.delete_many(query)

# Print the number of deleted documents
print(f"Number of documents deleted: {result.deleted_count}")
```

```python
# Check if any remaining documents include Dover
query = { 'LocalAuthorityName': 'Dover' }


documents = establishments.find(query)


if documents.count() > 0:
    print("There are remaining documents with 'Dover' in LocalAuthorityName.")
else:
    print("No remaining documents with 'Dover' in LocalAuthorityName.")
```

```python
# Check that other documents remain with 'find_one'

remaining_document = establishments.find_one()


if remaining_document:
    print("There are remaining documents in the collection.")
else:
    print("No remaining documents in the collection.")
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
filter = {}

# Define the update to change the data type of longitude and latitude
update = [
    { '$set': { 'geocode.longitude': { '$toDouble': '$geocode.longitude' } } },
    { '$set': { 'geocode.latitude': { '$toDouble': '$geocode.latitude' } } }
]

# Update the documents
result = establishments.update_many(filter, update)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
filter = { 'RatingValue': { '$ne': '' } }


update = [
    {
        '$set': {
            'RatingValue': {
                '$cond': {
                    'if': { '$ne': ['$RatingValue', ''] },
                    'then': { '$toInt': '$RatingValue' },
                    'else': '$RatingValue'
                }
            }
        }
    }
]
```

```python
# Check that the coordinates and rating value are now numbers

documents = establishments.find()


for document in documents:
    longitude = document['geocode']['longitude']
    latitude = document['geocode']['latitude']
    rating_value = document['RatingValue']
    
    if isinstance(longitude, (int, float)) and isinstance(latitude, (int, float)) and isinstance(rating_value, (int, float)):
        print("Coordinates and RatingValue are numbers.")
    else:
        print("Coordinates and/or RatingValue are not numbers.")
```
